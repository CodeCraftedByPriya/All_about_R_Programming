```r
# Plot
plot(1,8)
plot(c(1,8), c(3,4))
```

```r
x <- c(1, 5, 8, 7)
y <- c(5, 7, 6, 8)
plot(x,y)
```

```r
# Sequence
plot(4:21)
```

```r
# Draw a line & Axis names & Properties
plot(1:10, type ="l", main="My Graph", xlab="The x-axis", ylab="The y axis", col = "Red", lwd = 2, lty = 5)
# lwd ---> width of the line (Thickness)
# lty ---> Stype of the line. Ranges from 0 to 6
"Available parameter values for lty:
0 removes the line
1 displays a solid line
2 displays a dashed line
3 displays a dotted line
4 displays a \"dot dashed\" line
5 displays a \"long dashed\" line
6 displays a \"two dashed\" line"
```

```r
plot(4:21, col = "blue", cex = 2, main = "Properties of the dots", pch = 13)
# cex ---> size of the dots
# pch ---> Type of the dots. Ramnges from 0 to 25
```

```r
# Multiple Lines
line1 <- c(1,2,3,4,5,10)
line2 <- c(2,5,7,8,9,10)
plot(line1, type = "l", col = "blue")
lines(line2, type="l", col = "red")
```

```r
# Scatter Plots - compare
# day one, the age and speed of 12 cars:
x1 <- c(5,7,8,7,2,2,9,4,11,12,9,6)
y1 <- c(99,86,87,88,111,103,87,94,78,77,85,86)
# day two, the age and speed of 15 cars:
x2 <- c(2,2,8,1,15,8,12,9,7,3,11,4,7,14,12)
y2 <- c(100,105,84,105,90,99,90,95,94,100,79,112,91,80,85)
plot(x1, y1, main="Observation of Cars", xlab="Car age", ylab="Car speed", col="red", cex=2)
points(x2, y2, col="blue", cex=2)
```

```r
# Pie Charts
x1 <- c(20, 45, 12)
pie(x1)
```

```r
label1 <- c("Samsung", "Apple", "RealMe")
colours <- c("Blue", "Yellow", "Cyan")
pie(x1, init.angle = 90, label = label1, main = "Mobile MArket", col = colours)    # The start will be froma 90 deg angle
legend("bottomright", label1, fill = colours)
# The legend can be positioned as either: bottomright, bottom, bottomleft, left, topleft, top, topright, right, center
```

```r
# Bar Graph
x <- c("A", "B", "C", "D")
y <- c(2, 4, 6, 8)
barplot(y, names.arg = x, col = "Turquoise", density = 25, width = c(1, 2, 2, 2.5))     # names.arg defines the names of each observation in the x-axis
```

```r
# Horizontal Bars
barplot(y, names.arg = x, horiz = TRUE, col = "red", density = 10)
```
